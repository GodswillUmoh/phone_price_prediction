```python
import pandas as pd
#import numpy as np
```

```python
#load the dataset
phone_data = pd.read_csv("phone.csv")
phone_data
```

```python
#Data exploration
phone_data.describe()

```

```python
#Splitting data to X (six inputs needed for the prediction) and y (output)
y = phone_data["price_range"]
X = phone_data.drop(["pc", "price_range", "int_memory", "m_dep", "mobile_wt", "n_cores", "px_height", "px_width", "ram", "sc_h", "sc_w", "talk_time", "three_g", "touch_screen", "wifi"], axis=1)
X
```

```python
#Data splitting
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split (X, y, test_size=0.2)

X_train
```

```python
X_test
```

```python
#creating model for prediction
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

#for a battery_power of 840, it is expected to predict a medium cost of 1
predictions = model.predict([[840,	0,	2.2,	0,	1,	0	]])
predictions
```

```python
phone_data 
```

```python
#prediction of clock_speed being 0.4, it is expected to predict the High Score of 2
predictions_2 = model.predict(X_test)
predictions_2
```

```python
#4 Measuring the accuracy of each model
from sklearn.metrics import accuracy_score

score = accuracy_score(y_test, predictions_2)
score
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)

y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)

y_lr_train_pred

```

```python
#more model
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
y_rf_train_pred
```

```python
y_rf_test_pred 
```

```python
#Evaluate model performance, comparing actually value from the predicted value
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_trian_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)

print("LR MSE (Train)",lr_train_mse)
print("LR r2_score (Train)",lr_trian_r2)
print("LR MSE (Test)",lr_test_mse)
print("LR r2_score (Test)",lr_test_r2)
```

```python
#Measuring the accuracy of each model
from sklearn.metrics import accuracy_score
prediction_3 = model.predict(X_test)

score_2 = accuracy_score(y_test, prediction_3)
print(score_2)
```

```python
#Evaluating the performance of Random Forest Classifier

#from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_trian_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)

print("RF MSE (Train)",rf_train_mse)
print("RF r2_score (Train)",rf_trian_r2)
print("RF MSE (Test)",rf_test_mse)
print("RF r2_score (Test)",rf_test_r2)

"""As classification is more appropriate for 
discrete or distinct values, 
the mean_squared_error is closer to One which is 
a a good performance 
"""
```
